```python
import pandas as pd
import numpy as np
import ast
```

```python
#BxMS data
bx_board = pd.read_csv("BxMS_board.csv")
bx_sprint = pd.read_csv("BxMS_sprint.csv")
bx_issue = pd.read_csv("BxMS_issues.csv")

#Jboss
jb_board = pd.read_csv("JBoss_board.csv")
jb_sprint = pd.read_csv("JBoss_sprint.csv")
jb_issue = pd.read_csv("JBoss_issues.csv")

#KOGITO
ko_board = pd.read_csv("KOGITO_board.csv")
ko_sprint = pd.read_csv("KOGITO_sprint.csv")
ko_issue = pd.read_csv("KOGITO_issues.csv")

#WindUP
win_board = pd.read_csv("WINDUP_board.csv")
win_sprint = pd.read_csv("WINDUP_sprint.csv")
win_issue = pd.read_csv("WINDUP_issues.csv")

#Aurora
au_board = pd.read_csv("Aurora_board.csv")
au_sprint = pd.read_csv("Aurora_sprint.csv")
au_issue = pd.read_csv("Aurora_issues.csv")

#MESOS
me_board = pd.read_csv("MESOS_board.csv")
me_sprint = pd.read_csv("MESOS_sprint.csv")
me_issue = pd.read_csv("MESOS_issues.csv")

#Usergrid
us_board = pd.read_csv("USERGRID_board.csv")
us_sprint = pd.read_csv("USERGRID_sprint.csv")
us_issue = pd.read_csv("USERGRID_issues.csv")
```

### Join Table
#### Board

```python
bx_board = bx_board.iloc[: , 1:]
jb_board = jb_board.iloc[: , 1:]
ko_board = ko_board.iloc[: , 1:]
win_board = win_board.iloc[: , 1:]
me_board = me_board.iloc[: , 1:]
us_board = us_board.iloc[: , 1:]

board_list = [bx_board, jb_board,ko_board, win_board,au_board,me_board,us_board]
all_board = pd.concat(board_list)
all_board = all_board.reset_index()

all_board.drop(columns=all_board.columns[0], axis=1, inplace=True)
all_board.drop(columns=all_board.columns[2], axis=1, inplace=True)
```

```python
#Create subset of all nested columns
issue_links = all_issue[["board_id", "issue_id", "issue_key","issue_links"]]
issue_sprint = all_issue[["board_id", "issue_id", "issue_key", "issue_related_sprint"]]
issue_comment = all_issue[["board_id", "issue_id", "issue_key", "comment_content"]]
issue_fixed_version = all_issue[["board_id", "issue_id", "issue_key", "issue_fixed_version"]]
```

#### Flat neasted data
Issue links

```python
link = pd.DataFrame()
for i in issue_links.index:
    try:
        a = pd.json_normalize(ast.literal_eval(issue_links["issue_links"][i])) 
        a["board_id"] = issue_links['board_id'][i]
        a["issue_id"] = issue_links['issue_id'][i]
        a["issue_key"] = issue_links['issue_key'][i]
        frame = [link, a]
        link = pd.concat(frame)
    except:
        pass
```

```python
link = link[["board_id", "issue_id", "issue_key","type.id","type.inward","type.outward","outwardIssue.id","outwardIssue.key"]]
issue_links = link.dropna()
issue_links = issue_links.reset_index(drop = True)
issue_links = issue_links.rename(columns={"type.id": "link_type_id", "type.inward": "link_inward", 
                                        "type.outward":"link_inward","outwardIssue.id":"linked_issue_id", "outwardIssue.key":"linked_issue_key"}, errors="raise")
#issue_links.to_csv("IssueLinks.csv", index=False)
```

Issue related sprint

```python
sprint = pd.DataFrame()
for i in issue_sprint.index:
    try:
        a = pd.json_normalize(ast.literal_eval(issue_sprint["issue_related_sprint"][i])) 
        a["board_id"] = issue_sprint['board_id'][i]
        a["issue_id"] = issue_sprint['issue_id'][i]
        a["issue_key"] = issue_sprint['issue_key'][i]
        frame = [sprint, a]
        sprint = pd.concat(frame)
    except:
        pass
```

```python
sprint = sprint[["board_id", "issue_id", "issue_key","id","state","name","startDate","endDate","completeDate","activatedDate"]]
issue_sprint = sprint.dropna()
issue_sprint = issue_sprint.reset_index(drop = True)
issue_sprint= issue_sprint.rename(columns={"id": "sprint_id", "name": "sprint_name", "startDate":"start_date","endDate":"end_date","completeDate":"complete_date","activatedDate":"activated_date"}, errors="raise")
#issue_sprint.to_csv("IssueSprint.csv", index=False)
```

Issue Comment

```python
comment = pd.DataFrame()
for i in issue_comment.index:
    try:
        a = pd.json_normalize(ast.literal_eval(issue_comment["comment_content"][i])) 
        a["board_id"] = issue_comment['board_id'][i]
        a["issue_id"] = issue_comment['issue_id'][i]
        a["issue_key"] = issue_comment['issue_key'][i]
        frame = [comment, a]
        comment = pd.concat(frame)
    except:
        pass
```

```python
comment = comment[["board_id", "issue_id", "issue_key","body"]]
issue_comment = comment.dropna()
issue_comment = issue_comment.reset_index(drop = True)
issue_comment= issue_comment.rename(columns={"body": "comment_content"}, errors="raise")
issue_comment.to_csv("IssueComment.csv", index=False)
```

```python
fixed = pd.DataFrame()
for i in issue_fixed_version.index:
    try:
        a = pd.json_normalize(ast.literal_eval(issue_fixed_version["issue_fixed_version"][i])) 
        a["board_id"] = issue_fixed_version['board_id'][i]
        a["issue_id"] = issue_fixed_version['issue_id'][i]
        a["issue_key"] = issue_fixed_version['issue_key'][i]
        frame = [fixed, a]
        fixed = pd.concat(frame)
    except:
        pass
```

```python
fixed = fixed[["board_id", "issue_id", "issue_key","description","name"]]
issue_fixed_version = fixed.dropna()
issue_fixed_version = issue_fixed_version.reset_index(drop = True)
issue_fixed_version= issue_fixed_version.rename(columns={"description": "version_description", "name":"version_name"}, errors="raise")
issue_fixed_version.to_csv("IssueFixedVersion.csv", index=False)
```
